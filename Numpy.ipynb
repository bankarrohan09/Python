Computational efficiency

```python
def digits(n1):
  sum = 0
  while (n1>0):
    r = n1 % 10
    sum = sum + r
    n1 = n1 // 10
  print(sum)
```

```python
digits(9318786654)
```

```python
%timeit digits(64321103)
```
